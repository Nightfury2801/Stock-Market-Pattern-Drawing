```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Data/BPCL.csv')
temp_df =  df.head(100)
temp_df = temp_df[['Date','Close']]
temp_df.head()
```

```python

def directional_change(date, close: np.array, high: np.array, low: np.array, sigma: float):
    
    up_zig = True # Last extreme is a bottom. Next is a top. 
    tmp_max = high[0]
    tmp_min = low[0]
    tmp_max_i = 0
    tmp_min_i = 0

    tops = []
    bottoms = []

    for i in range(len(close)):
        if up_zig: # Last extreme is a bottom
            if high[i] > tmp_max:
                tmp_max = high[i]
                tmp_max_i = i
            elif close[i] < tmp_max - tmp_max * sigma: 
                top = [i, date[i-1], tmp_max]
                tops.append(top)

                # Setup for next bottom
                up_zig = False
                tmp_min = low[i]
                tmp_min_i = i
        else: # Last extreme is a top
            if low[i] < tmp_min:
                # New low, update 
                tmp_min = low[i]
                tmp_min_i = i
            elif close[i] > tmp_min + tmp_min * sigma: 
                bottom = [i, date[i-1], tmp_min]
                bottoms.append(bottom)

                # Setup for next top
                up_zig = True
                tmp_max = high[i]
                tmp_max_i = i

    return tops, bottoms

def get_extremes(ohlc: pd.DataFrame, sigma: float):
    tops, bottoms = directional_change(ohlc['Date'], ohlc['Close'], ohlc['Close'], ohlc['Close'], sigma)
    tops = pd.DataFrame(tops, columns=['i', 'Date', 'Close'])
    bottoms = pd.DataFrame(bottoms, columns=['i', 'Date', 'Close'])
    # tops['type'] = 1
    # bottoms['type'] = -1
    extremes = pd.concat([tops, bottoms])
    extremes = extremes.set_index('i')
    extremes = extremes.sort_index()
    return extremes

keypoint_sorted_df = get_extremes(temp_df, 0.03)
keypoint_sorted_df
```

```python
def localTopRollingWindow(temp_df, order):
    l_max = []
    n=temp_df.shape[0]
    for i in range(order,n-order):
        flag=0
        for j in range(1, order+1):
            if temp_df.iloc[i]['Close'] < temp_df.iloc[i+j]['Close'] or temp_df.iloc[i]['Close'] < temp_df.iloc[i-j]['Close']:
                flag=1
                break;
        if flag==0:
            l_max.append(temp_df.iloc[i])
    
    return pd.DataFrame(l_max)
```

```python
def localBottomRollingWindow(temp_df, order):
    l_min = []
    n=temp_df.shape[0]
    for i in range(order,n-order):
        flag=0
        for j in range(1, order+1):
            if temp_df.iloc[i]['Close'] > temp_df.iloc[i+j]['Close'] or temp_df.iloc[i]['Close'] > temp_df.iloc[i-j]['Close']:
                flag=1
                break;
        if flag==0:
            l_min.append(temp_df.iloc[i])
    return pd.DataFrame(l_min)
```

```python
keypoint_df = pd.concat([localTopRollingWindow(temp_df, 2), localBottomRollingWindow(temp_df, 2)], ignore_index = True)
keypoint_sorted_df = keypoint_df.sort_values('Date')
```

```python
plt.plot(temp_df['Date'], temp_df['Close'])
plt.plot(keypoint_sorted_df['Date'], keypoint_sorted_df['Close'])
plt.rc('xtick', labelsize=2)
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price closing')
plt
```
